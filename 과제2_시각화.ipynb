## 뇌혈관 질환 위험도 분류, 각 특징 파악 (성별, 연령) 

### 심뇌혈관질환이란? 


심뇌혈관질환은 동맥경화 진행에 의해 허혈성 심장질환(심근경색, 협심증 등)을 대표로 하며 고혈압, 부정맥, 판막질환, 심근증, 심부전 등의 심장질환과 뇌졸증(중풍)과 말초혈관질환 등의 혈관질환이 있다. 따라서 심뇌혈관질환은 심장질환과 기타 혈관질환을 모두 일컫는 매우 광범위한 질환이다. 

### 심뇌혈관질환의 원인과, 위험요소? 

동맥경화증이 주 원인. 심뇌혈관질환의 위험요소로는 고혈압, 당뇨병, 고지혈증, 흡연이 가장 대표적이고 그 ㅇ외에도 관상동맥질환의 가족력, 연령(남자 45세 이상, 여자 55세 이상), 비만, 심한 스트레스, 운동 부족, 여성의 폐경기 이후 등이 위험요인으로 작용 

### 심뇌혈관질환의 위험인자의 종류와 예방관리 

1. 담배는 끊기 
2. 술은 하루에 한 두잔 이하로 줄이기 
3. 비체질량지수 25kg/m^3 이상이면 고혈압, 당뇨병, 고지혈증의 위험이 2배 이상 증가하며, 30kg/m^3이상인 경우에는 각 질환의 사망률이 1.5배 증가함 
4. 고혈압 기준: 수축기 140mmHg 이상 또는 이완기 90mmHg 이상
5. 당뇨병 치료: 당뇨병이 있는 경우 위험이 2배 많게는 4배까지 높아지게 됨.

진단기준: 공복혈당이 126 이상인 경우 
당화혈색소 검사결과가 6.5% 이상인 경우 
6. 이상지혈증

#### 위험군2- 담배, 음주 둘 중에 하나라도 하는 사람 

```python
df_a_or_s1=  df[(df['흡연상태']=='Yes') & (df['음주여부'] == 'No')]
df_a_or_s2=  df[(df['흡연상태']=='No') & (df['음주여부'] == 'No')]
df_a_or_s=pd.concat([df_a_or_s1, df_a_or_s2])
```

```python
plt.rc('font', family='NanumBarunGothic')
f,ax=plt.subplots(figsize=(20,10))
sns.set_style('whitegrid')
sns.set(font='NanumBarunGothic')
g=sns.countplot(x='연령대코드(5세단위)', data=df_a_or_s, order=df['연령대코드(5세단위)'].value_counts(ascending=False).index, 
                hue='성별코드', palette= 'PuRd')

plt.title('담배, 음주 중에 하나라도 하는 집단의 연령 분포', fontsize=20)

plt.legend(loc = 'upper right')
plt.show()
```



*   위험군 1과 달리, 담배나 음주 둘 중에 하나라도 하는 사람을 추출하니, 여성의 비율이 압도적으로 많았다. 
*   연령대는 60-64세(13) > 50-54(11) > 55-59세(12) 순으로 많았다. 



#### 위험군3- 당뇨병과 고혈압 환자 분포 

```python
df_bp= df[(df['수축기혈압'] >=140) | (df['이완기혈압'] >=90)]
df_d= df[(df['식전혈당(공복혈당)'] >=126)]
```

```python
fig,ax=plt.subplots(nrows=2, figsize=(10,15))
ax[0].set_title("수축기 혈압과 체중", fontsize=14)
ax[1].set_title("이완기 혈압과 체중", fontsize=14)
sns.scatterplot(x='수축기혈압', y='체중(5Kg 단위)', hue='성별코드', data=df_bp,
                   palette='Purples', ax=ax[0])


sns.scatterplot(x='이완기혈압', y='체중(5Kg 단위)', hue='성별코드', data=df_bp,
                   palette='Purples', ax=ax[1])

```



*   수축기 혈압이 140이상인 경우에 고혈압으로 간주하므로, 140이상인 사람들의 몸무게 분포를 보면, 60~80kg에 가장 많이 몰려 있는 것을 확인할 수 있다. 이 중에는, 남자의 비율이 더 높다. 
*  이완기 혈압이 90 이상인 경우에 고혈압으로 간주하므로, 90 이상인 사람들의 몸무게 분포를 보면, 60~100kg까지 골고루 분포하는 것을 확인할 수 있다. 이완기 역시, 남자의 비율이 더 높다. 



```python
fig,ax=plt.subplots(figsize=(10,8))

ax=sns.scatterplot(x='식전혈당(공복혈당)', y='체중(5Kg 단위)', hue='성별코드', data=df_d,
                   palette='Purples') 

plt.title('공복혈당과 체중', fontsize=20)
```




*   당뇨병의 경우, 공복혈당이 126이상일 때, 당뇨병으로 간주한다. 밀집도로 보면, 55~70kg에 많이 분포하고 있는 것을 확인할 수 있다. 이들의 키와 허리둘레까지 고려하지 않았으니, 비만이라고 판정하기에는 무리가 있다. 




#### 위험군4- 이상지혈증 

```python
df_ch= df[(df['총콜레스테롤'] >=250) | (df['LDL콜레스테롤'] >=170) |(df['HDL콜레스테롤'] >=65) | 
          (df['트리글리세라이드'] >=135)]
df_ch
```

```python
import matplotlib.font_manager as fm 
font_list=fm.findSystemFonts()
```

```python
sns.set(font="NanumBarunGothic", 
        rc={"axes.unicode_minus":False}, style='whitegrid')

fig,ax=plt.subplots(nrows=2, ncols=2, figsize=(18,16))
sns.lineplot(x='시도코드', y='총콜레스테롤', data=df_ch, ax=ax[0,0])


sns.lineplot(x='시도코드', y='트리글리세라이드', data=df_ch, ax=ax[0,1])


sns.lineplot(x='시도코드', y='LDL콜레스테롤', data=df_ch, ax=ax[1,0])

sns.lineplot(x='시도코드', y='HDL콜레스테롤', data=df_ch, ax=ax[1,1])
ax[0,0].set_title("시도코드와 총콜레스테롤", fontsize=14)
ax[0,1].set_title("시도코드와 트리글리세라이드", fontsize=14)
ax[1,0].set_title("시도코드와 LDL 콜레스테롤", fontsize=14)
ax[1,1].set_title("시도코드와 HDL콜레스테롤", fontsize=14)

plt.show()
```





*   총콜레스테롤과 LDL의 경우, 그래프의 분포가 비슷하다. 총콜레스테롤의 경우, 시도코드가 30인 대전광역시에서 총콜레스테롤이 가장 낮게 나왔고, 시도코드가 36인 세종특별자치시에서 총콜레스테롤이 제일 높게 나왔다. LDL 그래프 역시 마찬가지이다. 
*   트리글리세라이드의 경우, 시도코드가 29(광주광역시), 46(전라남도)에서 가장 높게 나왔고 31(울산광역시)에서 가장 낮게 나왔다. 


*   HDL의 경우, 시도코드가 31인 울산에서 가장 높게 나왔고 46인 전라남도에서 가장 낮게 나왔다. 또한, 29, 30인 광주, 대전에서도 낮게 나왔다. 
*   트리글리세라이드와 HDL의 경우 광주와 전라남도의 경우 분포가 비슷한걸 알 수 있다. 지역별로 음식의 차이가 있으니, 이런 비슷한 결과가 나왔을 것이라고 추측해볼 수 있다.
